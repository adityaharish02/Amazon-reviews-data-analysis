```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_one_company(company, n):
    df_one_company = df_small.loc[df_small['BrandName'].isin([company])]
    pivot = pd.pivot_table(df_one_company,
            values = ['Rating', 'ReviewVotes'],
            index =  ['ProductName'],
                       columns= [],
                       aggfunc=[np.sum, np.mean, np.count_nonzero, np.std], 
                       margins=True, fill_value=0).sort_values(by=('count_nonzero', 'Rating'), ascending=False).fillna('')
    top_10_prods = pivot.reindex().head(n)
    top_10_prods = top_10_prods.reset_index()
    tt_prods = top_10_prods['ProductName']
    tt_prods2 = tt_prods.reset_index()
    top_10_prods_list = list(set(tt_prods2['ProductName']))
    #top_30_prod_list

    try:
        aa= df_one_company[df_one_company['ProductName'].isin(top_10_prods_list)]
        g = sns.factorplot(x='ProductName', 
                           y='Rating',
                           data=aa, 
                           saturation=1, 
                           kind="bar", 
                           ci=None, 
                           aspect=4, 
                           linewidth=1) 
        locs, labels = plt.xticks()
        plt.setp(labels, rotation=90)
    except: 
        pass
for i in top_10_brand_list:
    plot_one_company(i, 10)
```

#Relationship between helpfulness of review and length of review
Here we plot the average length of reviews and the average number of votes based on the helpfulness. Let’s see if more number of people find longer reviews more helpful. There is an acceptable positive correlation (r = 0.343) between the two as supported by the trend line below.

```python
from scipy.stats.stats import pearsonr

df_reviewVotes = df['ReviewVotes'].astype(int)
df_reviewL = df['Reviews'].fillna('').str.len().astype(int)
print 'Correlation coefficient = ', pearsonr(df_reviewVotes,df_reviewL)[0]
ax = sns.regplot(x=df_reviewL[:50000], y=df_reviewVotes[:50000])
ax.set(xlabel='Review length', ylabel='No. of review votes')
plt.show()
```

#Relationship between price and number of reviews
Explore correlation between product price and number of reviews. This will help us answer questions like: Do expensive products receive more number of reviews?
The scatter below says not necessarily. Statistically the correlation is negligible (r = -0.031). So there is no relationship between price and the number of reviews it gets.

```python
from collections import Counter
value = Counter(df['Price'].astype(float))
x_val = pd.Series(value.keys())
y_val = pd.Series(value.values())
print 'Correlation coefficient = ', pearsonr(x_val,y_val)[0]
ax = sns.regplot(x=x_val, y=y_val)
ax.set(xlabel='Price', ylabel='No. of reviews')
plt.show()
```

 #Relationship between product price and rating
Now find out if costlier products have better ratings. This plot shows there is negligible correlation (r = 0.076) between rating and price. When consumers pay more for a product, they also expect better quality and sellers need to meet this expectation. However it can be considered(BY LOOKING AT THE GRAPH) that with cost the product quality increases, which in turn leads to higher rating.

```python
from scipy.stats.stats import pearsonr
df_price = df['Price'].astype(float)
df_rating = df['Rating'].fillna('0.0').astype(float)
print 'Correlation coefficient = ', pearsonr(df_price,df_rating)[0]
ax = sns.regplot(x=df_price, y=df_rating)
ax.set(xlabel='Price', ylabel='Rating')
plt.show()
```

#Relationship between review length and price
Explore relationship between the length of the reviews and phone price. The plot shows that there is no increment in the length of the reviews with increase in the price.  The correlation is very close to zero. The correlation remains weak (r = 0.01).

```python
df_reviewL = df['Reviews'].fillna('').str.len().astype(int)
df_price = df['Price'].astype(float)
print 'Correlation coefficient = ', pearsonr(df_reviewL,df_price)[0]
ax = sns.regplot(x=df_reviewL[:100000], y=df_price[:100000])
ax.set(xlabel='Review Length', ylabel='Price')
plt.show()
```
